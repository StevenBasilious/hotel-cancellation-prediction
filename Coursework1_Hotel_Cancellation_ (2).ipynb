## 1.2 Removing Inconsistent values and Outliers (10%)

Detecting inconsistencies can be achieved through a variety of methods. Some can be identified by examining unique values within each column, while others may require a solid understanding of the problem domain. Since you might not be an expert in the hotel or hospitality industry, here are some helpful hints:

Hints:

1. Check for incomplete bookings, such as reservations with zero adults, babies, or children.
2. Examine rows with zeros in both 'stays_in_weekend_nights' and 'stays_in_week_nights.'



```python
# check the rows with no guests
no_guests = HotelData[(HotelData['adults'] == 0) &
                      (HotelData['children'] == 0) &
                      (HotelData['babies'] == 0)]
print("Bookings with no adults, children, and babies:", no_guests.shape[0])

# check the rows with negative guests
negatives = HotelData[(HotelData['adults'] < 0) |
                      (HotelData['children'] < 0) |
                      (HotelData['babies'] < 0)]
print("Rows with negative guest counts:", negatives.shape[0])

# check the booking with 0 nights
no_stay = HotelData[(HotelData['stays_in_weekend_nights'] == 0) &
                    (HotelData['stays_in_week_nights'] == 0)]
print("Bookings with zero stay length:", no_stay.shape[0])

# check the ADR (Average Daily Rate) with negative values
if 'adr' in HotelData.columns:
    negative_adr = HotelData[HotelData['adr'] < 0]
    print("Rows with negative ADR:", negative_adr.shape[0])
```

```python
# remove guests with 0 or negative values
HotelData = HotelData[~((HotelData['adults'] == 0) &
                        (HotelData['children'] == 0) &
                        (HotelData['babies'] == 0))]

# remove guests with negative values
HotelData = HotelData[(HotelData['adults'] >= 0) &
                      (HotelData['children'] >= 0) &
                      (HotelData['babies'] >= 0)]

# remove rows with 0 nights
HotelData = HotelData[~((HotelData['stays_in_weekend_nights'] == 0) &
                        (HotelData['stays_in_week_nights'] == 0))]

#remove the ADR with negative
HotelData = HotelData[HotelData['adr'] >= 0]



#comment lead time
HotelData = HotelData[HotelData['lead_time'] >= 0]

# Remove extreme outliers (top 1%)
upper_limit = HotelData['adr'].quantile(0.99)
HotelData = HotelData[HotelData['adr'] <= upper_limit]

#Remove long stays
HotelData = HotelData[(HotelData['stays_in_week_nights'] <= 30) &
                      (HotelData['stays_in_weekend_nights'] <= 10)]
print("Shape after cleaning:", HotelData.shape)
```

The main purpose for this code is to check and remove the unrealistic values like booking with zero guests (adults, children, or babies = 0) or negative count for guests, negative prices, stays with zero nights and extreme outliers Lead times below zero and very long stays (more than 30 weekdays or 10 weekend nights) were also removed. Also top 1% booking with high daily rates has been removed as they have extreme values and removing them will help the model to be to have more stabilize dataset and make sure that the model put more effort on normal booking patterns, so it ensure that the dataset have realistic hotel booking and to improve the quality of the model training

# 4. Classifier Training (20%)


---


Utilise the sklearn Python library to train a ML model (e.g.decision tree classifier). Your process should start with splitting your dataset into input features (X) and a target feature (y). Next, divide the data into 70% training and 30% testing subsets. Train your model on the training dataset and evaluate using test dataset with appropriate metrics. Aim to achieve higher accuracy e.g. more than 70% accuracy using your model.

## 4.1. Data Splitting (5%)

```python
# Define target variable and features
y = HotelData['is_canceled']

# Drop irrelevant or leakage columns
X = HotelData.drop(columns=['is_canceled', 'reservation_status', 'reservation_status_date'], errors='ignore')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print("Training data shape:", X_train.shape)
print("Testing data shape:", X_test.shape)
```

## 4.2. Model Training (10%)

```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

# Apply one-hot encoding to categorical features
X_train_encoded = pd.get_dummies(X_train, drop_first=True)
X_test_encoded = pd.get_dummies(X_test, drop_first=True)

# Initialize model
rf = RandomForestClassifier(
    n_estimators=200,        # number of trees
    max_depth=None,          # allow trees to expand fully
    random_state=42,
    n_jobs=-1,               # use all cores
    class_weight='balanced'  # handle class imbalance
)

# Train on the training data
rf.fit(X_train_encoded, y_train)
```

## 4.3. Model Evaluation (5%)

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# Make predictions on the test set
# Ensure test set columns match training set columns after one-hot encoding
X_test_encoded = X_test_encoded.reindex(columns=X_train_encoded.columns, fill_value=0)

y_pred = rf.predict(X_test_encoded)
y_prob = rf.predict_proba(X_test_encoded)[:, 1]

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_prob)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print(f"ROC AUC Score: {roc_auc:.4f}")

# Display classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Plot confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Canceled', 'Canceled'], yticklabels=['Not Canceled', 'Canceled'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

for the Classifier training section it has been used Random Forest Classifier from sklearn to train to predict if the hotel booking will be canceled or no.
in the first part (Data splitting ) the dataset was divided into two target variable y which is (is_canceld)  and x which is all other features that will be used to get the target y and the dataset split into training 70% and testing 30% by using train_test_split.

The classifier was set up using (n_estimators) using 200 decision trees to achieve consistent outcomes, (class_weight='balanced) which is for to locate any disparity in any class between the booking that canceled and not canceled, (random_state=42) for guarantee reproducibility and to optimize the training time the model was trained by using (X_train, y_train) data.
